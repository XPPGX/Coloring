```python
import numpy as np
import matplotlib.pyplot as plt
import networkx as nx
import pandas as pd

datasetDataFrame = pd.read_csv('/home/arthur/algorithm/Coloring/Sequential/karate.csv')
colorDataFrame = pd.read_csv('/home/arthur/algorithm/Coloring/Sequential/nodesColor.csv')
nodeNum = len(colorDataFrame)
edgeNum = len(datasetDataFrame)
print("nodeNum = {}".format(nodeNum))
print("edgeNum = {}".format(edgeNum))

# 初始化graph
g = nx.Graph()
# 取得nodeList
nodeList = colorDataFrame['node'].values.tolist()
g.add_nodes_from(nodeList)
print("# nodes: " + str(g.number_of_nodes()))

# 取得edgeList
edgeList = datasetDataFrame.values.tolist()
g.add_edges_from(edgeList)
print("# edges: " + str(g.number_of_edges()))

# 取得node的colorList
colorIndexList = colorDataFrame['color'].values.tolist()
print(colorIndexList)
unique, counts = np.unique(colorIndexList, return_counts=True)
print("顏色出現次數 : ", dict(zip(unique, counts)))

nx.draw_networkx(g, with_labels=True, node_color = colorIndexList)
```
